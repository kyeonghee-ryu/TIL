## 2. Visualization & EDA

```python
plt.figure(figsize = (20,15))
sns.heatmap(train_df.corr(), annot = True, fmt = ' .2f', square = True)
plt.show()
```

```python
sns.catplot(data=train_df, x = 'OverallQual', y='SalePrice', aspect=2)
```

```python
sns.regplot(train_df['GrLivArea'], train_df['SalePrice'])
```

```python
train_df.plot.scatter(x='TotalBsmtSF', y='SalePrice', ylim =(0,800000) )
```

```python
sns.catplot(kind = 'box', data=train_df, x = 'GarageCars', y='SalePrice', aspect=2)
```

```python
sns.distplot(train_df['SalePrice'])
```

```python
print("Skewness: %f" % train_df['SalePrice'].skew())  #왜도
print('Kurtosis: %f' % train_df['SalePrice'].kurtosis()) #첨도
```
